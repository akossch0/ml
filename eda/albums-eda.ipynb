This looks erronous, we will have to check which track is actually in our datase to decide which row to keep out of the two.

```python
# load experimental/liked_songs_with_audio_features_and_genres.json
df_songs = pd.read_json('../experimental/liked_songs_with_audio_features_and_genres.json')
# select 3Q1DprZauQyNo8KGlY8brZ and 3JcvpTfI64SmXxUl7XUvQI ids
df_songs = df_songs[df_songs['id'].isin(['3Q1DprZauQyNo8KGlY8brZ', '3JcvpTfI64SmXxUl7XUvQI'])]
df_songs
```

We see two added dates, one recent, and one from 2 years ago. I specifically remember liking this song 2 years ago, and recently relistening to it, after which I removed it from the liked to re-like it again so that it is moved up to the top of the liked songs list. However, between these two times it might have happened that Spotify reuploaded the song, it was assigned a different id, but the entry for the previous id was also kept. This is because instead of modifying the playlists of each individual user having this song on their playlist, they solve this issue by forwarding the old id to the new id. We validated this by typing
spotify:track:3Q1DprZauQyNo8KGlY8brZ   
into our browser opening the Spotify player, and then checking the id of the song that actually played.
https://open.spotify.com/track/3JcvpTfI64SmXxUl7XUvQI?si=448153ae45de451d
It is clear what is happenning.

We will delete the older entry from both the albums and liked songs.

```python
# remove album with 5yHn5NJqFHuRscn44o9POq id and reindex the DataFrame
df = df[df['id'] != '5yHn5NJqFHuRscn44o9POq']
df.reset_index(drop=True, inplace=True)
```

## Release date column

```python
import pandas as pd

def preprocess_dates(date_str):
    if len(date_str) == 4 and date_str.isdigit():
        return date_str + '-01-01'
    else:
        return date_str

df['release_date'] = df['release_date'].apply(preprocess_dates)

df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
```

```python
year_buckets = [5, 10, 20]

plot_by_year_buckets(df, 'release_date', year_buckets)
```

```python
# select album with highest release date
df[df['release_date'] == df['release_date'].max()]
```

```python
# select album with lowest release date
df[df['release_date'] == df['release_date'].min()]
```
